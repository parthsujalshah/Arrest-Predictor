```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

### Data Cleaning

```python
# since we already have latitude and longitude, we can drop x_coordinate and y_coordinate column
df = df.drop(['X Coordinate', 'Y Coordinate'], axis=1)
```

```python
df = df.dropna()
```

```python
# IUCR, Community Area, FBI Code, and Primary Type are categorical
columns = ['IUCR', 'Community Area', 'FBI Code', 'Primary Type', 'Arrest', 'Domestic', 'Year']
for col in columns:
    df[col] = df[col].astype("category")
```

```python
df.drop_duplicates(keep='first')
```

```python
df['Date'] = pd.DatetimeIndex(df['Date'])
```

```python
df['Date']
```

```python
df.columns
```

```python
# selecting important varibales
selected_cols = ['Arrest', 'Primary Type', 'Domestic', 'Community Area', 'Year']
```

```python
df = df[selected_cols]
```

```python
df
```

```python
# creating dummy variables
df = pd.get_dummies(df)
```

```python
df
```

### Model Building and Optimization of Hyperparameters

```python
X = df.drop(['Arrest_True', 'Arrest_False'], axis=1)
```

```python
Y = df['Arrest_True']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

### Optimization of Hyperparameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_para = {'criterion':['gini','entropy'],'max_depth':[10,20,30,40]}
```

```python
gs = GridSearchCV(DecisionTreeClassifier(), tree_para)
```

```python
gs.fit(X_train, y_train)
```

```python
print(gs.best_estimator_)
```

### Model Building

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=30)
```

```python
dt.fit(X_train, y_train)
```

### Model Evaluation

```python
Y
```

```python
y_train
```

```python
for i in X_test.columns:
    print(i)
```

```python
y_predict = dt.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_predict)
```

```python
confusion_matrix(y_predict, y_test)
```

```python
print(classification_report(y_test, y_predict))
```
